```python
# feature tables
TA = pd.read_csv('MT02_SRM/full_Feature_table.tsv', sep='\t', header=0, index_col=0)
TA
```

```python
TX = pd.read_csv('MT02_SRM/MT02_XCMS_featureTable.csv', sep=',', header=0, index_col=0)
TX
```

```python
NIST_TA = TA[['mz', 'rtime', 'batch14_MT_20210808_005']][TA['batch14_MT_20210808_005']>0]
NIST_TA
```

```python
NIST_TX = TX[['mzmed', 'rtmed', 'batch14_MT_20210808_005.mzML']][TX['batch14_MT_20210808_005.mzML'].notna()]
NIST_TX
```

```python
# match functions
# 0.000010 is 10 ppm
PPM_tolerance = 0.000005
# use a large number to include anything in RTime, small number to be specific
RTime_tolerance = 10       # seconds in retention time, usually a small number
                            # more lenient for diff instruments
                            # and possible diff void volume
             
# F1 = (m/z, rt)
def match2(F1, F2, PPM_tolerance=PPM_tolerance, RTime_tolerance=RTime_tolerance):
    if abs(F1[0]-F2[0])/F1[0] < PPM_tolerance and abs(F1[1] - F2[1]) < RTime_tolerance:
        return True
    else:
        return False
    
# test
match2((129.1541, 55), (129.1538, 14))
```

```python
mz_a, rt_a = NIST_TA['mz'].to_list(), NIST_TA['rtime'].to_list()
mz_x, rt_x = NIST_TX['mzmed'].to_list(), NIST_TX['rtmed'].to_list()
N1_, N2_ = len(mz_a), len(mz_x)
print(N1_, N2_)
```

```python
good = []
for ii in range(N1_):
    F1 = (mz_a[ii], rt_a[ii])
    for jj in range(N2_):
        F2 = (mz_x[jj], rt_x[jj])
        if match2(F1, F2):
            good.append((ii, jj))
```

```python
# not in overlap but in XCMS
set2 = set([x[1] for x in good])
print(len(set2))
xcms_only = [jj for jj in range(N2_) if jj not in set2]
print(len(xcms_only), xcms_only[:5])
```

```python
# not in overlap but in asari
set1 = set([x[0] for x in good])
print(len(set1))
asari_only = [ii for ii in range(N1_) if ii not in set1]
print(len(asari_only), asari_only[:5])
```

```python
good_xcms = [x[1] for x in good]
use_pairs = [x for x in good if good_xcms.count(x[1])==1]

g2 = [x[0] for x in good]
use_pairs2 = [x for x in good if g2.count(x[0])==1]

print(len(use_pairs2), len(use_pairs))

LL = set(use_pairs2).intersection(set(use_pairs))

print(len(LL))
```

```python

X0 = [x[0] for x in LL]
X1 = [x[1] for x in LL]
v_xcms = NIST_TX['batch14_MT_20210808_005.mzML'].iloc[X1].to_list()
v_asari = NIST_TA['batch14_MT_20210808_005'].iloc[X0].to_list()


CM = pd.DataFrame(
{
    'xcms': v_xcms, 'asari': v_asari
})

print(CM.shape)
CM = CM[CM['xcms'] > 1][CM['asari']>1]
print(CM.shape)

CM = np.log2(CM+1)

CM.corr()
```

```python
plt.figure(figsize=(4,4))
plt.plot(CM['xcms'], CM['asari'], 'r.' )
plt.title("common top features in XCMS and asari in NIST SRM1950")

plt.savefig("SRM_batch14_MT_20210808_005_compr.pdf")
```
